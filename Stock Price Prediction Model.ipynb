```python
import yfinance as yf #yfinance is an API for yahoo finance to use in jupyter notebook
```

```python
sp500=yf.Ticker('^GSPC') #TIcker is a class that calls any  specific index(company) from the website. Here, GSPC is any random sp500 index
```

```python
sp500=sp500.history(period='max') #history is a method that will query all the data in the data rame
```

```python
sp500
```

```python
sp500.index
```

```python
sp500.plot.line(y='Close', use_index=True) #Draw a line graph using the close parameter in the dataset
```

```python
# Delete unnecessary columns for the dataset
del sp500['Dividends']
del sp500['Stock Splits']
sp500
```

```python
#Add a new column to show the data of tomorrow using shift method 
sp500['Tomorrow']=sp500['Close'].shift(-1)
sp500
```

```python
#Add a new column for our target variable which gives a boolean/integer value.
sp500['Target']=(sp500['Tomorrow']> sp500['Close']).astype(int)
sp500
```

```python
#Remove the data before 1990
sp500=sp500.loc['1990-01-01':].copy()
sp500
```

```python
#Run an initial Model & train_test
from sklearn.ensemble import RandomForestClassifier
m1=RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train=sp500.iloc[:-100]
test=sp500.iloc[-100:]

#Columns used to predict data
predictors=['Open', 'Close', 'Volume', 'High', 'Low']
m1.fit(train[predictors], train['Target']) 
```

```python
#See metrics of your model
from sklearn.metrics import precision_score
pred=m1.predict(test[predictors])
pred
```

```python
import pandas as pd
preds=pd.Series(pred,index=test.index)
preds
```

```python
#Score of model
precision_score(test['Target'], preds)
```

```python
#Combine the actual data with predicted data
combined=pd.concat([test['Target'], preds],axis=1)
combined.plot()
```

```python
#Build a back testing system for the model

def predict(train, test, predictors, m1):
    m1.fit(train[predictors], train['Target'])
    pred=m1.predict(test[predictors])
    pred=pd.Series(pred, index=test.index, name='Predictions')
    combine=pd.concat([test['Target'], pred], axis=1)
    return combine
```

```python
def backtest(data,model, predictors,start=2500, step=250):
      all_predictions=[]

      for i in range(start, data.shape[0], step):
        train=data.iloc[0:i].copy()
        test=data.iloc[i:(i+step)].copy()
        predictions=predict(train,test,predictors, m1)
        all_predictions.append(predictions)

      return pd.concat(all_predictions)
```

```python
predictions=backtest(sp500, m1, predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
#New Score
precision_score(predictions['Target'], predictions['Predictions'])
```

```python
predictions['Target'].value_counts()/predictions.shape[0]
```

```python
#Add more predictors to modelto improve accuracy
horizon=[2,5,60,1000] #These are days of rolling averages means the data of these days will be analysed
new_predictors=[]

for horizons in horizon:
    rolling_avg=sp500.rolling(horizons).mean()

    ratio_column=f"Close_Ratio_{horizons}"
    sp500[ratio_column]=sp500['Close'] / rolling_avg['Close']

    trend_column=f"Trend{horizons}"
    sp500[trend_column] = sp500.shift(1).rolling(horizons).sum()['Target']

    new_predictors +=[ratio_column, trend_column]
```

```python
sp500=sp500.dropna()
sp500
```

```python
#Update model
m2=RandomForestClassifier(n_estimators=100, min_samples_split=50, random_state=1)
```

```python
#Update predit function
def predict(train, test, predictors, m2):
    m2.fit(train[predictors], train['Target'])
    pred=m2.predict_proba(test[predictors])[:,1] #Adds probability of price going up or down of any row to be 0 or 1
    pred[pred>=.6]=1 # custom threshold of chances for pricce to go up. 0. means 60 %
    pred[pred<.6] =0
    pred=pd.Series(pred, index=test.index, name='Predictions')
    combine=pd.concat([test['Target'], pred], axis=1)
    return combine
```

```python
predictions = backtest(sp500, m2, new_predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions['Target'], predictions['Predictions'])
```
